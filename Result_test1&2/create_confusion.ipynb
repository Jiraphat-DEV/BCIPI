```python
import pandas as pd

# Load the Excel file to check available sheet names
file_path = './ผลการทดลอง BCI EYE.xlsx'

# Load the Excel file
excel_file = pd.ExcelFile(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import classification_report
import os

# Example label mapping (replace with your actual label mapping if needed)
label_mapping = {
    "Original Value": ['Left', 'Center', 'Right']
}

for i in range(1, 26):
    Subject = i
    text_S = f"Subject_{Subject}"

    # Load the data from "Subject_1" sheet to inspect its content and locate the confusion matrix
    subject_1_df = pd.read_excel(file_path, sheet_name=text_S)

    confusion_matrix_1 = subject_1_df.iloc[[5, 6, 7], [5, 6, 7]]
    confusion_matrix_2 = subject_1_df.iloc[[13, 14, 15], [5, 6, 7]]

    # Create directory for saving results
    folder_name = os.path.join(f"results_S{Subject}")
    os.makedirs(folder_name, exist_ok=True)

    def plot_confusion_matrix(cm, cm_filename, labels):
        """Plot and save confusion matrix heatmap."""
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.title('Confusion Matrix - Test Set')
        plt.savefig(cm_filename)
        plt.close()

    def generate_classification_report(cm, report_filename, labels):
        """Generate and save classification report from confusion matrix."""
        y_true, y_pred = [], []

        cm = np.array(cm)
        for i in range(cm.shape[0]):
            for j in range(cm.shape[1]):
                y_true.extend([i] * cm[i, j])  # True class i
                y_pred.extend([j] * cm[i, j])  # Predicted class j

        report = classification_report(y_true, y_pred, target_names=labels, digits=4)
        
        with open(report_filename, "w") as f:
            f.write(report)

    def process_confusion_matrix(cm, matrix_name, labels):
        """Process confusion matrix by plotting and generating report."""
        cm_filename = os.path.join(folder_name, f"{matrix_name}.png")
        report_filename = os.path.join(folder_name, f"{matrix_name}_report.txt")
        
        # Plot confusion matrix and save
        plot_confusion_matrix(cm, cm_filename, labels)
        
        # Generate classification report and save
        generate_classification_report(cm, report_filename, labels)

    # Process confusion matrix 1
    process_confusion_matrix(confusion_matrix_1.astype(int), "confusion_matrix_1", label_mapping["Original Value"])

    # Process confusion matrix 2
    process_confusion_matrix(confusion_matrix_2.astype(int), "confusion_matrix_2", label_mapping["Original Value"])

    print("Confusion matrices and classification reports have been saved.")
```
